```python
def locateAge(inAge):
    age=int(inAge)
    LocatedAge='Age 30 to 34'
    if 0<=age<=4 : LocatedAge='Age 0 to 4'
    elif 5<=age<=9 : LocatedAge='Age 5 to 9'
    elif 10<=age<=14 : LocatedAge='Age 10 to 14'
    elif 15<=age<=19 : LocatedAge='Age 15 to 19'
    elif 20<=age<=24 : LocatedAge='Age 20 to 24'
    elif 25<=age<=29 : LocatedAge='Age 25 to 29'
    elif 30<=age<=34 : LocatedAge='Age 30 to 34'
    elif 35<=age<=39 : LocatedAge='Age 35 to 39'
    elif 40<=age<=44 : LocatedAge='Age 40 to 44'
    elif 45<=age<=49 : LocatedAge='Age 45 to 49'
    elif 50<=age<=54 : LocatedAge='Age 50 to 54'
    elif 55<=age<=59 : LocatedAge='Age 55 to 59'
    elif 60<=age<=64 : LocatedAge='Age 60 to 64'
    elif 65<=age<=69 : LocatedAge='Age 65 to 69'
    elif 70<=age<=74 : LocatedAge='Age 70 to 74'
    elif 75<=age<=79 : LocatedAge='Age 75 to 79'
    elif 80<=age<=84 : LocatedAge='Age 80 to 84'
    elif 85<=age : LocatedAge='Age 85 and over'    
    return LocatedAge
```
